### Slide: V2G Optimization Using MILP for Peak Shaving

**Title:** V2G Optimization Using MILP for Peak Shaving

**Content:**

#### Introduction
- **Brief Overview:**
  - Explain that MILP (Mixed-Integer Linear Programming) is used to optimize the charging and discharging schedules of EVs to achieve peak shaving.

#### Objective Function
- **Objective:**
  - Minimize the peak load on the grid by scheduling EV charging and discharging.

#### Details:
- **Formulation:**
  - **Minimize:** \( \text{max\_load} \)
  - Where \(\text{max\_load}\) is the maximum load observed during the day.
  
- **Variables:**
  - **Charging profiles:** \( \text{chargeprofiles}_{i,t} \)
    - Amount of power (kW) an EV \(i\) charges at time \(t\).
  - **Discharging profiles:** \( \text{dischargeprofiles}_{i,t} \)
    - Amount of power (kW) an EV \(i\) discharges at time \(t\).
  - **Total load:** \( \text{total\_load}_t \)
    - Total grid load at time \(t\).
  - **State variables:** \( \text{chargestates}_{i,t} \) and \( \text{dischargestates}_{i,t} \)
    - Binary variables indicating whether an EV \(i\) is charging or discharging at time \(t\).

- **Constraints:**
  - **Load balancing:**
    \[
    \text{total\_load}_t = \text{base\_load}_t + \sum_{i} (\text{chargeprofiles}_{i,t} - \text{dischargeprofiles}_{i,t})
    \]
    - Ensures the total load includes base load and net EV load (charging minus discharging).
  - **Charging constraints:**
    \[
    0 \leq \text{chargeprofiles}_{i,t} \leq \text{chargestates}_{i,t} \times \text{power\_managed\_Upper}
    \]
    - EV charging power is within allowed limits when charging.
  - **Discharging constraints:**
    \[
    \text{power\_V2G\_Lower} \leq \text{dischargeprofiles}_{i,t} \leq \text{dischargestates}_{i,t} \times \text{power\_V2G\_Upper}
    \]
    - EV discharging power is within allowed limits when discharging.
  - **State constraints:**
    \[
    \text{chargestates}_{i,t} + \text{dischargestates}_{i,t} \leq 1
    \]
    - Ensures an EV cannot charge and discharge simultaneously.
  - **State of Charge (SoC):**
    \[
    0 \leq \text{SoC}_{i,t} \leq 1
    \]
    - Ensures the battery state of charge remains within feasible limits.

#### Summary
- **Purpose:**
  - Balance the grid load by leveraging the flexibility of EVs to charge during low demand periods and discharge during high demand periods.
- **Benefits:**
  - Reduces peak load, enhances grid stability, and can defer investments in grid infrastructure.

**Content:**

- **Brief Overview:**
  - Explain that while peak shaving benefits the grid, it may not always be economically favorable for EV users.

#### How Peak Shaving Can Harm EV Users
- **Explanation:**

- **Higher Costs:**

  - **Increased Bills:** Charging during these times can significantly increase the overall electricity bills for EV users.

### Slide: Objective 1 - Minimum Load Variance

**Title:** Objective 1: Minimum Load Variance

**Content:**
- **Goal:** Stabilize the grid by minimizing the fluctuations in the daily load curve.
- **Method:**
  - **Mean Square Deviation:** Used to characterize load fluctuations.
  - **Strategy:** EVs discharge during peak load periods and charge during low load periods.
  - **Objective Function:** Minimize the variance of the daily load curve.

**Formula:**
\[ \text{Minimize} \quad \sigma^2 = \frac{1}{24} \sum_{t=1}^{24} (L_t - \bar{L})^2 \]
- Where \( L_t \) is the load at time \( t \) and \( \bar{L} \) is the average load over 24 hours.

### Slide: Objective 2 - Maximum Revenue for EV Users

**Title:** Objective 2: Maximum Revenue for EV Users

**Content:**
- **Goal:** Maximize the profits for EV users by optimizing charging and discharging times.
- **Method:**
  - **Price Fluctuations:** Charge when prices are low, discharge when prices are high.
  - **Battery Degradation Costs:** Consider the cost due to battery wear and tear.
  - **Objective Function:** Maximize the daily profits of EV users.

**Formula:**
\[ \text{Maximize} \quad \sum_{t=1}^{24} [r_p(t) \cdot P_{\text{ev}}(t) - r_b \cdot P_{\text{ev}}(t)] \]
- Where \( r_p(t) \) is the price at time \( t \), \( P_{\text{ev}}(t) \) is the power of EV at time \( t \), and \( r_b \) is the battery degradation cost.

**Details:**
- **Charge/Discharge Strategy:**
  - \( P_{\text{ev}}(t) \geq 0 \): Charging, \( \lambda_c = 1 \), \( \lambda_d = 0 \)
  - \( P_{\text{ev}}(t) \leq 0 \): Discharging, \( \lambda_c = 0 \), \( \lambda_d = 1 \)
- **Battery Degradation Cost:**
  - \( r_b = \frac{M}{\xi \cdot k} \)
  - \( r_b = 0.62 \) yuan/kWh
  - \( E_{\text{ev}} = 60 \) kW, \( k = 3000 \) cycles, \( \xi = 0.8 \)
